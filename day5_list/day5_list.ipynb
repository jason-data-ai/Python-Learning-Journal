````python
# ìì¬ ë‚©í’ˆ ë¦¬ìŠ¤íŠ¸ (ê³µê¸‰ì‚¬, ìì¬ëª…, ë‹¨ê°€)
items = [
    ("Aìƒì‚¬", "ì² ê·¼", 3000),
    ("Bìì¬", "ì‹œë©˜íŠ¸", 4300),
    ("Aìƒì‚¬", "ì² ê·¼", 3100),
    ("Cìì¬", "ëª¨ë˜", 800),
    ("Bìì¬", "ì² ê·¼", 2900),
    ("Aìƒì‚¬", "ì‹œë©˜íŠ¸", 4400),
    ("Cìì¬", "ì² ê·¼", 3200)
]

# ìš”êµ¬ì‚¬í•­
# 1. ìì¬ëª…ì´ 'ì² ê·¼'ì¸ ê²ƒë§Œ ê³¨ë¼ë‚´ê³ 
steel = {}
for item in items:    # ì—¬ê¸°ì„œëŠ” ì¡°ê±´ì„ ì˜ ë§ì¶°ì•¼ í•˜ë‚˜ì˜ í‚¤ì— ë¦¬ìŠ¤íŠ¸ í˜•íƒœë¡œ ë³µìˆ˜ì˜ ê°’ì´ ë“¤ì–´ê°ˆ ìˆ˜ ìˆì–´ì„œ ì¡°ê±´ì„ ì‹ ì¤‘íˆ ì„¸ì› ì–´.
    if item[1] == "ì² ê·¼" and not item[0] in steel:
        steel[item[0]] = [item[2]]
    elif item[1] == "ì² ê·¼" and item[0] in steel:
        steel[item[0]].append(item[2])

# 2. ê³µê¸‰ìë³„ë¡œ ë‹¨ê°€ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ëˆ„ì 
unit_price = []
for company in steel:    # ì´ê±´ ë¬¸ì œê°€ ê¼­ ë¦¬ìŠ¤íŠ¸ì— ëˆ„ì í•˜ë¼ê³  í•´ì„œ íšŒì‚¬ì™€ ë‹¨ê°€ëŠ” ë³´í˜¸ë˜ì•¼ í•œë‹¤ê³  ìƒê° -> íŠœí”Œ í˜•íƒœë¡œ ì €ì¥í–ˆì–´.
    unit_price.append((company, steel[company]))

# 3. ê° ê³µê¸‰ìì˜ í‰ê·  ê°€ê²©ì„ ê³„ì‚°í•˜ì—¬ ë”•ì…”ë„ˆë¦¬ë¡œ ì¶œë ¥
# â†’ ìµœì¢… ê²°ê³¼ ì˜ˆì‹œ: {'Aìƒì‚¬': 3050.0, 'Bìì¬': 2900.0, 'Cìì¬': 3200.0}
avg_company = {}
for company in unit_price:    # ë‹¨ê°€ê°€ ë¦¬ìŠ¤íŠ¸ì˜ í˜•íƒœë¡œ ë˜ì–´ìˆê¸° ë•Œë¬¸ì— í‰ê· ì„ êµ¬í•˜ë ¤ë©´ ë¦¬ìŠ¤íŠ¸ì˜ ê¸¸ì´ê°€ í•„ìš”í•´ì„œ qtyë¼ëŠ” ë³€ìˆ˜ë¥¼
    qty = len(company[1])     # ë§Œë“¤ì—ˆê³ , í•©ê³„ë¥¼ ë°”ë¡œ ì´ ê¸¸ì´ë¡œ ë‚˜ëˆ„ì–´ì„œ í•´ê²°í–ˆì–´. ê·¸ë¦¬ê³  ì†Œìˆ˜ì  ì´í•˜ í•œìë¦¬ í˜•ì‹ìœ¼ë¡œ ë§ì·„ì–´.
    avg_company[company[0]] = round(sum(company[1]) / qty, 1)
print(avg_company)
````
